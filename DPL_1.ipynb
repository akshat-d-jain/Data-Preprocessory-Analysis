```python
import numpy as np
import pandas as pd
import pickle as pk
def load_csv(filepath):
  data= []
  col= []
  checkcol=False
  with open("filepath") as f:
      for val in f.readlines():
          val=val.replace("\n","")
          val=val.split(',')
          if checkcol is False:
            col=val
            checkcol=True
          else:
            data.append(val)
  dfe =pd.Dataframe(data=data,columns=col)
  return dfe
```
