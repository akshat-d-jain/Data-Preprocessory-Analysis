```python
import pandas as pd
# to generate unique id
import uuid
import random
from faker import Faker
import datetime
```

Number of rows

```python
num_users=100
```

Add Feature List
id,gender,subscriber,name,rating

```python
features=[
    "id",
    "gender",
    "subscriber",
    "name",
    "rating"
]

df=pd.DataFrame(columns=features)

df
```

Generate Imbalanced Data
Generate unique identifiers

```python
df['id']=[uuid.uuid4().hex for i in range(num_users)]
df
```

Checking if all IDs are unique

```python
print(df['id'].nunique()==num_users)
```

Generating Gender values

```python
genders=["male","female","na"]

df['gender']=random.choices(
    genders,
    weights=(45,50,5),
    k=num_users
)
```

```python
df
```

```python
choice=[True,False]

df['subscriber']=random.choices(
    choice,
    k=num_users
)
```

```python
df
```

```python
#instanting faker
faker=Faker(locale="EN_IN")

def name_gen(gender):

  if gender=='male':
    return faker.name_male()
  elif gender=='female':
    return faker.name_female()

  return faker.name()

df['name']=[name_gen(i) for i in df['gender']]

df
```

```python
rating=[1,2,3,4,5]

df['rating']=random.choices(
    rating,
    weights=(5,10,10,10,5),
    k=num_users
)
df
```

```python
df.to_csv('fake_dataset.csv')
```
