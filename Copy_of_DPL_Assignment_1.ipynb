# Task 2: Data Cleaning
Dataset: Zomato missing value dataset

```python
import numpy as np
import pandas as pd
from sklearn.impute import KNNImputer
```

Reading the Datasets

```python
df=pd.read_csv('/content/zomato_dataset.csv')
df=pd.DataFrame(df)
df.head()
```

As we can see the there are no null values but we can see there are ? so we have to map the ? with np.nan which is using null values


Finding the missing data

```python
mapping={'?':np.nan}
df.replace({'Restaurant ID':mapping,'Restaurant Name':mapping,'Country Code':mapping,
             'City':mapping,'Address':mapping,'Locality':mapping,
             'Locality Verbose':mapping,'Longitude':mapping,'Latitude':mapping,
             'Cuisines':mapping,'Average Cost for two':mapping,'Currency':mapping,
             'Price range':mapping,'Aggregate rating':mapping,'Rating color':mapping,
             'Rating text':mapping,'Votes':mapping},inplace=True)
```

```python
df.isnull().sum()
```

Now due to mapping we can see there are some missing values containing in the zomato dataset
# Imputing the Dataset which contains numerical features


```python
df.head()
```

Numerical Data are country code,latitude,longitude,aggregrate rating,votes we will apply knn imputer on the following data till 5 nearest neighbour

```python
features_to_impute = ['Restaurant ID', 'Country Code','Longitude','Latitude','Aggregate rating','Votes','Price range']
```

```python
imputer = KNNImputer(n_neighbors=5)
```

```python
imputed_features = imputer.fit_transform(df[features_to_impute])
imputed_df = pd.DataFrame(imputed_features, columns=features_to_impute)
non_imputed_columns = [col for col in df.columns if col not in features_to_impute]
for col in non_imputed_columns:
    imputed_df[col] = df[col]
```

```python
imputed_df.head()
```

```python
imputed_df.isnull().sum()
```

```python
imputed_df['City'].fillna(method='bfill',inplace=True)
imputed_df['Address'].fillna(method='ffill',inplace=True)
imputed_df['Restaurant Name'].fillna(method='ffill',inplace=True)
imputed_df['Locality'].fillna(method='bfill',inplace=True)
imputed_df['Locality Verbose'].fillna(method='ffill',inplace=True)
imputed_df['Locality'].fillna(method='bfill',inplace=True)
imputed_df['Cuisines'].fillna(method='ffill',inplace=True)
imputed_df['Rating color'].fillna(method='bfill',inplace=True)
imputed_df['Currency'].fillna(method='ffill',inplace=True)
imputed_df['Rating text'].fillna(method='bfill',inplace=True)
imputed_df['Average Cost for two'].fillna(method='ffill',inplace=True)
```

```python
imputed_df.isnull().sum()
```

# A.Min-Max scaling


```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
numeric_columns = imputed_df.select_dtypes(include=['float64', 'int64']).columns
min_max_scaler = MinMaxScaler()
data_minmax_scaled = min_max_scaler.fit_transform(imputed_df[numeric_columns])
data_minmax_scaled = pd.DataFrame(data_minmax_scaled, columns=numeric_columns)
```

```python
data_minmax_scaled.head()
```

# 2.Z score scaling

```python
z_scaler = StandardScaler()
data_zscore_scaled = z_scaler.fit_transform(imputed_df[numeric_columns])
data_zscore_scaled = pd.DataFrame(data_zscore_scaled, columns=numeric_columns)
```
