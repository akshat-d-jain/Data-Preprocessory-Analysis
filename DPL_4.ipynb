K-NN Algorithm is The K Nearest Neighbour algorithm it is used to find null values,and we will use knn imputar with the help of scikit learn

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import KNNImputer
```

```python
dict={'Maths':[85,91,np.nan,91],
     'Chemistry':[72,75,74,np.nan],
     'Physics':[np.nan,68,82,81],
     'Biology':[81,81,82,np.nan]}
```

```python
Before_Imputation=pd.DataFrame(dict)
Before_Imputation
```

```python
imputer=KNNImputer(n_neighbors=2)
After_Imputation=imputer.fit_transform(Before_Imputation)
After_Imputation
```
