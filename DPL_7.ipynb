```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_fire=pd.read_csv('/content/Brazilian-fire-dataset.csv')
```

```python
df_fire.isnull().sum()
```

```python
order_col=['Date Reported','Year','State','Month','Number of Fires']
df_fire=df_fire[order_col]
df_fire.head()
```

```python
df_fire.describe(include='all')
```

```python
max_count1=df_fire['Number of Fires'].max()
df_fire[df_fire['Number of Fires']==max_count1]
```

```python
df_grp_fire=df_fire.groupby(['State']).agg({'Number of Fires':np.sum})
df_grp_fire.head(10)
```

```python
df_grp_fire.max()
```

```python
df_grp_fire.idxmax()
```

```python
df_grp_fire.idxmin()
```

```python
print("Max of Total Fires recorded were %d, and this was for the state:%s"\
      %(df_grp_fire['Number of Fires'].max(),\
        df_grp_fire.idxmax()[0]))

print("Min of Total Fires recorded were %d, and this was for the state:%s"\
      %(df_grp_fire['Number of Fires'].min(),\
        df_grp_fire.idxmin()[0]))
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(50,15))
sb.barplot(x='Year',y='Number of Fires',data=df_fire)
plt.show()
```

```python
fig=plt.figure(figsize=(40,30))
sb.barplot(x=df_grp_fire.index,y='Number of Fires',data=df_grp_fire)
```

```python
fig=plt.figure(figsize=(40,30))
sb.barplot(x=df_grp_fire.index,y='Number of Fires',data=df_grp_fire)
plt.xlabel('State',fontsize=30)
plt.ylabel('Number of fire recorded',fontsize=30)
plt.xticks(fontsize=25)
plt.yticks(fontsize=20)
plt.title(label='Total number of fires vs State',fontsize=50)
plt.show()
```
